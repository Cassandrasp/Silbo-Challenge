```python
import sqlite3
import pandas as pd
import PrendasDB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
PrendasDB.crearDB()
```

```python
PrendasDB.crear_prendas()
```

```python
print("Prendas creadas:")
prendas = PrendasDB.leer_prendas()
for prenda in prendas:
    print(prenda)
```

```python
query = '''
        SELECT * 
        FROM StockPrendas
        '''
df = PrendasDB.sql_query(query)
```

```python
df = df.drop(columns='id_prenda')
df
```

```python
prendas = df['prenda'].tolist()
tallas = df['talla'].tolist()
colores = df['color'].tolist()
```

```python
datos = {
     "prenda": prendas,
     "talla": tallas,
     "color": colores
}
```

```python
# Convertir los datos a un DataFrame de pandas
df_knn = pd.DataFrame(datos)

knn = KNeighborsClassifier(n_neighbors=3)
```

```python

# Copiamos el dataframe original para no modificar los datos originales
df_numerico = df_knn.copy()

# Convertimos las columnas 'talla' y 'color' a valores numéricos
label_encoder_talla = LabelEncoder()
df_numerico['talla'] = label_encoder_talla.fit_transform(df_numerico['talla'])

label_encoder_color = LabelEncoder()
df_numerico['color'] = label_encoder_color.fit_transform(df_numerico['color'])

# Mostramos el DataFrame resultante
print(df_numerico.head())

# Seleccionamos las características y etiquetas
X_numerico = df_numerico[['talla', 'color']]
y_numerico = df_numerico['prenda']

# Entrenamos el modelo KNN con las características numéricas
knn.fit(X_numerico, y_numerico)
```

```python

# Creamos una función para convertir un string en valores numéricos
def convertir_string_a_numerico(string_prenda, string_talla, string_color):
    talla_numerico = label_encoder_talla.transform([string_talla])[0]
    color_numerico = label_encoder_color.transform([string_color])[0]
    return talla_numerico, color_numerico

# Convertimos el string a valores numéricos
talla_num, color_num = convertir_string_a_numerico("Camiseta", "M", "Azul")

# Predecimos la prenda más cercana usando el modelo KNN
prenda_predicha = knn.predict([[talla_num, color_num]])

# Imprimimos la prenda predicha
print("La prenda predicha es:", prenda_predicha[0])
```
